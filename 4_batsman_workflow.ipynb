```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# create state

class BatsmanState(TypedDict):
    runs : int
    balls_played : int
    fours : int
    sixes : int
    strike_rate : float
    balls_per_boundry : float
    boundry_percent : float
    summary : str
```

```python
def strike_rate(state : BatsmanState) -> BatsmanState:
    runs = state['runs']
    balls = state['balls_played']
    strike_rate = (runs / balls)*100

    return {'strike_rate' : strike_rate}
```

```python
def balls_per_boundry(state : BatsmanState) -> BatsmanState:
    balls = state['balls_played']
    fours = state['fours']
    sixes = state['sixes']
    balls_per_boundry = balls / (fours+sixes)

    return {'balls_per_boundry' : balls_per_boundry}
```

```python
def boundry_percent(state : BatsmanState) -> BatsmanState:
    boundry_percent = (((state['fours'] * 4) + (state['sixes'] * 6)) / state['runs']) * 100
    return {'boundry_percent' : boundry_percent}
```

```python
def summary(state : BatsmanState):
    summary = f"""
    Strike Rate -> {state['strike_rate']}\n
    Balls per Boundry -> {state['balls_per_boundry']}\n
    Boundry Percentage -> {state['boundry_percent']}
"""
    
    return {'summary' : summary}
```

```python
# create graph

graph = StateGraph(BatsmanState)

# add node
graph.add_node('strike_rate', strike_rate)
graph.add_node('balls_per_boundry', balls_per_boundry)
graph.add_node('boundry_percent', boundry_percent)
graph.add_node('summary', summary)

# add edges
graph.add_edge(START, 'strike_rate')
graph.add_edge(START, 'balls_per_boundry')
graph.add_edge(START, 'boundry_percent')

graph.add_edge('strike_rate', 'summary')
graph.add_edge('balls_per_boundry', 'summary')
graph.add_edge('boundry_percent', 'summary')

graph.add_edge('summary', END)

# compile graph
workflow = graph.compile()
```

```python
initial_state = {'runs' : 69, 'balls_played' : 35, 'fours' : 4, 'sixes' : 3}
final_state = workflow.invoke(initial_state)
print(final_state)
```
