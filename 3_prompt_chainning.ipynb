```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()

model = ChatOpenAI()
```

```python
# creating state

class BlogState(TypedDict):
    topic : str
    outline : str
    blog : str
```

```python
# def llm_blog(state : BlogState) -> BlogState:
#     topic = state['topic']
#     prompt1 = f"Create a outline to write a blog for this topic -> {topic}"
#     outline = model.invoke(prompt1).content
#     state['outline'] = outline
#     prompt2 = f"Write a blog on {topic} using following outline \n {outline}"
#     blog = model.invoke(prompt2)
#     state['blog'] = blog

#     return state
```

```python
def llm_outline(state : BlogState) -> BlogState:
    topic = state['topic']
    prompt1 = f"Create a outline to write a blog for this topic -> {topic}"
    outline = model.invoke(prompt1).content
    state['outline'] = outline

    return state
```

```python
def llm_blog(state : BlogState) -> BlogState:
    topic = state['topic']
    outline = state['outline']
    prompt2 = f"Write a blog on {topic} using following outline \n {outline}"
    blog = model.invoke(prompt2).content
    state['blog'] = blog

    return state
```

```python
# create graph
graph = StateGraph(BlogState)

# add nodes
graph.add_node('llm_outline', llm_outline)
graph.add_node('llm_blog', llm_blog)

# add edges
graph.add_edge(START, 'llm_outline')
graph.add_edge('llm_outline', 'llm_blog')
graph.add_edge('llm_blog', END)

# compile graph
workflow = graph.compile()
```

```python
initial_state = {'topic' : 'Indian Heritage'}
final_state = workflow.invoke(initial_state)
print(final_state)
```
